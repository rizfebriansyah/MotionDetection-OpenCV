```python
import cv2

#first video - Traffic_Laramie_1.mp4

#reads the input video
videoCap = cv2.VideoCapture('Traffic_Laramie_1.mp4')

#initialise OpenCV - Background Subtractor for KNN and MOG2
BS_KNN = cv2.createBackgroundSubtractorKNN()
BS_MOG2 = cv2.createBackgroundSubtractorMOG2()

cars = 0
while videoCap.isOpened():
    ret, frame = videoCap.read() #this will read the next frame

    #this is to extract the foreground mask
    forgMask = BS_MOG2.apply(frame)
    
 
    #this is to draw and display the reference lines
    cv2.line(frame,(300,300), (300,450), (34,34,178), 2) #RED LINE (centre)
    cv2.line(frame,(290,300), (290,450), (112,25,25), 1) #NAVY LINE (exit from left)
    cv2.line(frame,(310,300), (310,450), (112,25,25), 1) #NAVY LINE (entry from right)
    
    #this is to extract the contours
    conts, _ = cv2.findContours(forgMask, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    
    for c in conts:
        
        #ignore the small-sized contours (this will not create a bounding rectangle if countour area is less than 200)
        if cv2.contourArea(c) < 200:
            continue
        #however, if contour area is more than 200, bounding rectangle will be drawn
        x, y, w, h = cv2.boundingRect(c)
        if y > 336 and y < 363 and x > 202 and x < 219:
        
            #draw the bounding rectangle for all contours
            cv2.rectangle(frame,(x,y), (x+w, y+h), (0,255,0), 2)
            #this is to find the centre coordinate of the bounding rectangle, and then displaying the circle
            xMid = int((x + (x+w)) / 2) 
            yMid = int((y + (y+h)) / 2)
            cv2.circle(frame, (xMid, yMid), 2, (71,99,255), 5)
            #cars counter
            if xMid > 294 and xMid < 302:
                cars += 1
    
    #show the foreground mask and original video
    cv2.imshow('Foreground Mask', forgMask)
    cv2.putText(frame, 'CARS COUNTER: {}'.format(cars), (50, 200), cv2.FONT_HERSHEY_DUPLEX, 3, (0,225,255), 3)
    cv2.imshow('Original Video', frame)
    
    #wait for any key to be pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

#this is to release video capture
cv2.destroyAllWindows()
videoCap.release()
```
